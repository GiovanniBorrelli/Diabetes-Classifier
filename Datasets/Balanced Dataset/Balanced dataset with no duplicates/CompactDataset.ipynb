```python
import pandas as pd
```

```python
dataset = pd.read_excel("DiabeteBalanced.xlsx")
```

```python
# Crea una colonna "gender" unificata
dataset["gender"] = dataset[["gender_Female", "gender_Male", "gender_Other"]].idxmax(axis=1).str.split("_").str[-1]

# Crea una colonna "smoking_history" unificata
smoking_cols = ["smoking_history_No Info", "smoking_history_current", "smoking_history_ever",
                "smoking_history_former", "smoking_history_never", "smoking_history_not current"]

def get_smoking_history(row):
    for col in smoking_cols:
        if row[col] == 1:
            return col.split("_")[-1]
    return "No Info"

dataset["smoking_history"] = dataset.apply(get_smoking_history, axis=1)

# Rimuovi le colonne originali "gender_Female", "gender_Male", "gender_Other",
# "smoking_history_No Info", "smoking_history_current", "smoking_history_ever",
# "smoking_history_former", "smoking_history_never", "smoking_history_not current"
dataset.drop(columns=["gender_Female", "gender_Male", "gender_Other"] + smoking_cols, inplace=True)

print(dataset.head())

dataset.to_excel("DiabetesDataset.xlsx", index=False)
```
