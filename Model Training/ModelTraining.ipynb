```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score
```

```python
dataset = pd.read_excel("DiabeteBalanced.xlsx")
```

```python
attributes = ["age", "hypertension", "heart_disease", "bmi", "HbA1c_level",
"blood_glucose_level", "gender_Female", "gender_Male", "gender_Other", "smoking_history_No Info",
"smoking_history_current", "smoking_history_ever", "smoking_history_former",
"smoking_history_never", "smoking_history_not current"]
X = dataset[attributes]
y = dataset["diabetes"]

X = pd.get_dummies(X)

classifier = RandomForestClassifier()

scores_accuracy = cross_val_score(classifier, X, y, cv=5, scoring='accuracy')
scores_precision = cross_val_score(classifier, X, y, cv=5, scoring='precision')
scores_recall = cross_val_score(classifier, X, y, cv=5, scoring='recall')
scores_f1 = cross_val_score(classifier, X, y, cv=5, scoring='f1')

print("RANDOM FOREST CLASSIFIER - Metriche di valutazione:")
print(f"Accuracy: {scores_accuracy.mean():.2f}")
print(f"Precision: {scores_precision.mean():.2f}")
print(f"Recall: {scores_recall.mean():.2f}")
print(f"F1-Score: {scores_f1.mean():.2f}")
```

```python
classifier = DecisionTreeClassifier(random_state=42)

scores_accuracy = cross_val_score(classifier, X, y, cv=5, scoring='accuracy')
scores_precision = cross_val_score(classifier, X, y, cv=5, scoring='precision')
scores_recall = cross_val_score(classifier, X, y, cv=5, scoring='recall')
scores_f1 = cross_val_score(classifier, X, y, cv=5, scoring='f1')

print("DECISION TREE - Metriche di valutazione:")
print(f"Accuracy: {scores_accuracy.mean():.2f}")
print(f"Precision: {scores_precision.mean():.2f}")
print(f"Recall: {scores_recall.mean():.2f}")
print(f"F1-Score: {scores_f1.mean():.2f}")
```

```python
classifier = GaussianNB()

scores_accuracy = cross_val_score(classifier, X, y, cv=5, scoring='accuracy')
scores_precision = cross_val_score(classifier, X, y, cv=5, scoring='precision')
scores_recall = cross_val_score(classifier, X, y, cv=5, scoring='recall')
scores_f1 = cross_val_score(classifier, X, y, cv=5, scoring='f1')

print("NAIVE BAYES - Metriche di valutazione:")
print(f"Accuracy: {scores_accuracy.mean():.2f}")
print(f"Precision: {scores_precision.mean():.2f}")
print(f"Recall: {scores_recall.mean():.2f}")
print(f"F1-Score: {scores_f1.mean():.2f}")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
```

```python
classifier = RandomForestClassifier()

predicted_labels = cross_val_predict(classifier, X, y, cv=5)

conf_matrix = confusion_matrix(y, predicted_labels)

class_labels = ['Negativo', 'Positivo']

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='BuPu', fmt='d', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Etichette previste')
plt.ylabel('Etichette effettive')
plt.title('Matrice di Confusione - Random Forest')

plt.show()
```

```python
classifier = DecisionTreeClassifier(random_state=42)

predicted_labels = cross_val_predict(classifier, X, y, cv=5)

conf_matrix = confusion_matrix(y, predicted_labels)

class_labels = ['Negativo', 'Positivo']

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='BuPu', fmt='d', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Etichette previste')
plt.ylabel('Etichette effettive')
plt.title('Matrice di Confusione - Decision Tree')

plt.show()
```

```python
classifier = GaussianNB()

predicted_labels = cross_val_predict(classifier, X, y, cv=5)

conf_matrix = confusion_matrix(y, predicted_labels)

class_labels = ['Negativo', 'Positivo']

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='BuPu', fmt='d', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Etichette previste')
plt.ylabel('Etichette effettive')
plt.title('Matrice di Confusione - Bayes')

plt.show()
```
