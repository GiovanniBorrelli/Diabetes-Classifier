```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset = pd.read_excel("DiabetesDataset.xlsx")
```

```python
attributes = ["age", "hypertension", "heart_disease", "bmi", "HbA1c_level",
              "blood_glucose_level", "gender", "smoking_history"]
X = dataset[attributes]
y = dataset["diabetes"]

# Trasforma le variabili categoriche in variabili dummy
encoder = OneHotEncoder(drop="first", sparse=False)
X_encoded = pd.get_dummies(X, columns=["gender", "smoking_history"], drop_first=True, sparse=False)
encoded_columns = X_encoded.columns

model = RandomForestClassifier(random_state=42)
model.fit(X_encoded, y)
print("Modello addestrato.")

def predict_diabetes(gender, age, hypertension, heart_disease, smoking_history, bmi, HbA1c_level, blood_glucose_level):
    input_data = pd.DataFrame([[age, hypertension, heart_disease, bmi, HbA1c_level,
                                blood_glucose_level, gender, smoking_history]],
                              columns=attributes)

    input_encoded = pd.get_dummies(input_data, columns=["gender", "smoking_history"], drop_first=True, sparse=False)

    input_encoded = input_encoded.reindex(columns=encoded_columns, fill_value=0)

    prediction = model.predict(input_encoded)

    return prediction[0]
```

```python
input_data = {
    "gender": "Female",
    "age": 60,
    "hypertension": 0,
    "heart_disease": 0,
    "smoking_history": "current",
    "bmi": 30.5,
    "HbA1c_level": 9,
    "blood_glucose_level": 260,
}

result = predict_diabetes(**input_data)
if result == 0:
    print("Il paziente non ha il diabete.")
else:
    print("Il paziente ha il diabete.")
```
