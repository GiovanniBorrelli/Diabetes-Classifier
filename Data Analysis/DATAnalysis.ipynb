```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Diabete.xlsx')

genre_distribution = data['gender'].value_counts()

genre_distribution = genre_distribution.drop('Other')

colors = ['pink', '#6495ED'] #blu fiordaliso

# Crea il grafico
plt.bar(genre_distribution.index, genre_distribution.values, color=colors)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Male and Female')
plt.show()
```

```python
# Crea l'istogramma
plt.hist(data['age'], bins=30, color='#FFDB58', edgecolor='black', linewidth=0.5, rwidth=0.8)

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')

plt.show()
```

```python
hypertension_distribution = data['hypertension'].value_counts()

colors = ['#826AED', '#C879FF']
labels = ['No Hypertension', 'Hypertension']

plt.pie(hypertension_distribution, labels=labels, colors=colors, autopct='%1.1f%%',
        startangle=90, wedgeprops={'edgecolor': 'black'})

plt.axis('equal')  # Per rendere il grafico a torta circolare

plt.title('Distribution of Hypertension')

plt.show()
```

```python
heart_disease_distribution = data['heart_disease'].value_counts()

colors = ['salmon', 'skyblue']
labels = ['No Heart Disease', 'Heart Disease']

plt.bar(labels, heart_disease_distribution, color=colors)

plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.title('Distribution of Heart Disease')

plt.show()
```

```python
smoking_history_distribution = data['smoking_history'].value_counts()

colors = ['mistyrose', 'peachpuff', 'lightcoral', 'lightgreen', 'lightcyan', 'lightblue']
labels = smoking_history_distribution.index

plt.barh(labels, smoking_history_distribution, color=colors)

plt.xlabel('Count')
plt.ylabel('Smoking History')
plt.title('Distribution of Smoking History')

plt.show()
```

```python
# Seleziono gli attributi escludendo 'smoking_history', 'gender' e 'diabetes'
# (Perché non hanno valori numerici, e 'diabetes' in correlazione con sé stesso sarà sempre = 1.00)
selected_attributes = ['age', 'hypertension', 'heart_disease', 'bmi', 'HbA1c_level', 'blood_glucose_level']
df_selected = data[selected_attributes + ['diabetes']]

# Calcola la matrice di correlazione
correlation_matrix = df_selected.corr()

# Crea la matrice
sns.set(font_scale=0.7)  # Imposta la dimensione del carattere
sns.heatmap(correlation_matrix, annot=True, cmap='GnBu', fmt='.2f')

plt.title('Matrice di Correlazione')
plt.show()
```

```python
data = pd.read_excel('Diabete.xlsx')

attributes = ['age', 'hypertension', 'heart_disease', 'bmi', 'HbA1c_level', 'blood_glucose_level']
subset = data[attributes].copy()  # Utilizza .copy() per creare una copia del DataFrame

# Aggiungi la colonna target "diabetes"
subset['diabetes'] = data['diabetes']

# Calcola la matrice
correlation_matrix = subset.corr()

# Non mostrare la correlazione diabete-diabete
correlation_matrix = correlation_matrix.drop('diabetes')

# Crea la matrice
sns.set(font_scale=1.2)  # Imposta la dimensione del carattere
sns.heatmap(correlation_matrix[['diabetes']], annot=True, cmap='GnBu', fmt='.2f')

plt.title('Matrice di Correlazione - Diabete')
plt.show()
```

```python
attributes = ['gender', 'smoking_history']
subset = data[attributes].copy()

# Escludi 'Other' in 'gender' (solo 18 su 100000)
subset = subset[subset['gender'] != 'Other']

subset_encoded = pd.get_dummies(subset)

subset_encoded['diabetes'] = data['diabetes']

correlation_matrix = subset_encoded.corr()

# Non mostrare la correlazione diabete-diabete
correlation_matrix = correlation_matrix.drop('diabetes')

# Crea la matrice
sns.set(font_scale=1.2)
sns.heatmap(correlation_matrix[['diabetes']], annot=True, cmap='Oranges', fmt='.2f')

plt.title('Matrice di Correlazione - Diabete')
plt.show()
```
